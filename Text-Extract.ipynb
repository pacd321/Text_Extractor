```python
IMAGE_PATH_3 = 'quotes.png'
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(IMAGE_PATH_3)
result
text_only = [item[1] for item in result]
full_text = ' '.join(text_only)

print("Extracted text:")
print(full_text)
```

```python
target_language = 'hi'

translated = translator.translate(full_text, dest=target_language)

print(f"\nTranslated to {target_language}:")
print(translated.text)
```

```python
img = cv2.imread(IMAGE_PATH_3)
spacer = 100
for detection in result: 
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    img = cv2.putText(img,text,(20,spacer), font, 0.5,(0,255,0),2,cv2.LINE_AA)
    spacer+=15
    
plt.imshow(img)
plt.show()
```
